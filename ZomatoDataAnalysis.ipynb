```python
cols_to_drop = [
    'res_id', 'url', 'address', 'city_id', 'country_id',
    'zipcode', 'timings', 'opentable_support', 'currency'
]
```

```python
zomato = zomato.drop(columns=cols_to_drop, errors='ignore')
```

```python
zomato
```

```python
zomato.isna().sum()
```

```python
zomato['delivery'] = zomato['delivery'].replace(-1, np.nan)
zomato['takeaway'] = zomato['takeaway'].replace(-1, np.nan)
```

```python
zomato['cuisines'] = zomato['cuisines'].fillna('Not specified')
zomato['highlights'] = zomato['highlights'].fillna('Not specified')
```

```python
zomato['aggregate_rating'] = zomato['aggregate_rating'].replace(0, np.nan)
```

```python
zomato
```

```python
zomato.isna().sum()
```

```python
# Delivery / Takeaway flags: Yes / No / Unknown
zomato['delivery_flag'] = zomato['delivery'].map({1: 'Yes', 0: 'No'}).fillna('Unknown')
zomato['takeaway_flag'] = zomato['takeaway'].map({1: 'Yes', 0: 'No'}).fillna('Unknown')

# Price category (readable)
price_map = {1: 'Cheap', 2: 'Moderate', 3: 'Expensive', 4: 'Luxury'}
zomato['price_category'] = zomato['price_range'].map(price_map).fillna('Unknown')
```

```python
zomato
```

```python
zomato['establishment'] = zomato['establishment'].str.strip().str.title()
zomato['locality'] = zomato['locality'].str.strip()
zomato['name'] = zomato['name'].str.strip()
```

```python
zomato['primary_cuisine'] = zomato['cuisines'].apply(lambda x: x.split(',')[0].strip())
```

```python
est_freq = zomato['establishment'].value_counts(normalize=True)
common_est = est_freq[est_freq >= 0.01].index
zomato['establishment_grouped'] = zomato['establishment'].where(zomato['establishment'].isin(common_est), 'Other')
```

```python
top_cuisines = zomato['primary_cuisine'].value_counts().nlargest(15).index
zomato['cuisine_grouped'] = zomato['primary_cuisine'].where(zomato['primary_cuisine'].isin(top_cuisines), 'Other')
```

```python
zomato
```

```python
# Saving the cleaned dataframe

cols = [
  'name', 'city', 'locality', 'latitude', 'longitude',
  'establishment_grouped', 'cuisine_grouped',
  'price_category', 'average_cost_for_two',
  'aggregate_rating', 'rating_text', 'votes',
  'photo_count', 'delivery_flag', 'takeaway_flag',
    'highlights'
]

zomato[cols].to_csv('zomato_dashboard_cleaned.csv', index=False)
```
